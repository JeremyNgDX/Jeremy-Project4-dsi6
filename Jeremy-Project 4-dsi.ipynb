```python
company_list = []
job_title_list = []
salary_list = []
emp_type_list = []

for i in range(len(job_url1)):
    driver.get(job_url1.iat[i,0])
    sleep(3)
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    company_name = soup.find('p',{'class':'f6 fw6 mv0 black-80 mr2 di ttu'})
    job_title = soup.find('h1',{'class':'f3 fw6 mv0 pv0 mb1 dark-pink w-100 dib'})
    salary = soup.find('span',{'class':'salary_range dib f2-5 fw6 black-80'})
    location = soup.find('a',{'class':'link dark-pink underline-hover'})
    emp_type = soup.find('div',{'class':'db mt3 mb4'})
    company_list.append(company_name)
    job_title_list.append(job_title)
    salary_list.append(salary)
    emp_type_list.append(emp_type)

    print(company_name.text)
    print(job_title.text)
    print(salary.text)
    print(location.text)
    print(emp_type.text)
    print(len(company_list))
    print(len(job_title_list))
    print(len(salary_list))
    print(len(emp_type_list))
    
# NEED TO WRITE IN TRY EXCEPT TO LET PROGRAM CONTINUE RUNNING IN THE CASE 
# OF MISSING DATA
```
